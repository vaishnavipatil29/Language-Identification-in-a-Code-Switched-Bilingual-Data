```python
import pandas as pd
df = pd.read_csv('sentence_wise_wer.csv')
df
```

```python
org = []
# Python code to convert string to list character-wise
def Convert(string):
    list1=[]
    list1[:0]=string
    return list1
```

```python
org = []
pred = []

for i in range(1047):
    l1 = Convert(df['original'][i])
    l1.remove(' ')
    l2 = Convert(df['predicted'][i])
    l2.remove(' ')
    #print(l1)
    #print(l2)
    min_len = min(len(l1),len(l2))
    l1 = l1[0:min_len]
    l2 = l2[0:min_len]
    org.append(l1)
    pred.append(l2)
    
import itertools    

    
org_val = list(itertools.chain.from_iterable(org)) 
pred_val = list(itertools.chain.from_iterable(pred))
```

```python
from sklearn.metrics import classification_report
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

def print_classification_report(preds, y_test):

    # classification report
    #print(classification_report(y_test, preds))
    #333333333333df = pd.DataFrame(classification_report(y_test, preds,output_dict=True)).transpose()
    fig, ax = plt.subplots(figsize=(8,8))

    # color map
    cmap='Blues'
    # cmap = sn.diverging_palette(0, 230, 90, 60, as_cmap=True)

    cm = confusion_matrix(y_test, preds, labels=list(set(y_test)))
    # display(cm)
    df_cm = pd.DataFrame(cm, list(set(y_test)), list(set(y_test)))
    # display(df_cm)
    sn.set(font_scale=1.6)
    ax = sn.heatmap(df_cm, annot=True, annot_kws={"size": 15}, fmt='.4g', ax=ax, square=True)
    
    #ax = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, 
     #       fmt='.2%', cmap='Blues')

    ax.set_title('Telugu Confusion Matrix \n\n');
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ');

    ## Ticket labels - List must be in alphabetical order
    ax.xaxis.set_ticklabels(list(set(y_test)))
    ax.yaxis.set_ticklabels(list(set(y_test)))

    plt.show()
    return 
```

```python
print_classification_report(pred_val,org_val)
```
